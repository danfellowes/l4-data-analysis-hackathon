# Exploring the spread of COVID in Nigeria
### February 19th - May 19th 2020

I'll be exploring the initial spread of COVID in Nigeria by looking at number of cases per state over time, and comparing with total population data for each state, to attempt to discover whether the spread of COVID throughout the country started in one localised region or all at once across the country.

## Cases by State

```python
# add count column to sum
nga_cases['count'] = 1

# replace 'Abuja' with 'Federal Capital Territory' to match map data state names
nga_cases['province/state'][nga_cases['province/state'] == 'Abuja'] = 'Federal Capital Territory'
```

```python
# select columns & group by date + state
state_case_totals = (
    tp.from_pandas(nga_cases)
    .select('province/state', 'count')
    .summarize(total_cases = tp.sum(col('count')),
               by = 'province/state')
)

state_case_totals = state_case_totals.to_pandas()
state_case_totals_plot = state_case_totals.sort_values('total_cases', ascending=False).head(10)
```

```python
(ggplot(state_case_totals_plot, aes('province/state', 'total_cases'))
 + geom_col(aes(fill = 'total_cases'))
 + ggtitle('Total COVID cases in Nigeria by state')
 + geom_smooth()
 + geom_text(aes(label = 'total_cases'), nudge_y = 100)
 + labs(y = 'Total Cases', x = 'State', fill = 'Total Cases')
 + theme_minimal()
 + theme(axis_text_x = element_text(rotation=90),
         plot_title = element_text(size=18),
        figure_size = (12, 5)))
```

## Total cases by date

```python
# group and summarise with tp
case_dates = (
    tp.from_pandas(nga_cases)
    .summarize(total_cases = tp.sum(col('count')),
               by = 'date')
)

# convert back to pandas and convert to date
cases_df = case_dates.to_pandas()
cases_df['date'] = pd.to_datetime(cases_df['date'])
```

```python
(ggplot(cases_df, aes('date', 'total_cases'))
 + geom_col(aes(fill = 'total_cases'))
 + ggtitle('Total Cases by Date')
 + geom_smooth()
 + labs(x = 'Date', y = 'Total Cases', fill = 'Total Cases')
 + theme_minimal()
 + theme(axis_text_x = element_text(rotation=90),
        figure_size = (12, 5)))
```

## State cases by date

```python
# select columns & group by date + state
state_case_dates = (
    tp.from_pandas(nga_cases)
    .select('date', 'province/state', 'count')
    .summarize(total_cases = tp.sum(col('count')),
               by = ['date', 'province/state'])
)

state_case_dates = state_case_dates.to_pandas()

state_case_dates_plot = state_case_dates.rename(columns = {'province/state':'state'})
state_case_dates_plot['date'] = pd.to_datetime(state_case_dates_plot['date'])
```

```python
(ggplot(state_case_dates_plot, aes('date', 'total_cases'))
 + geom_smooth(color = 'blue')
 + geom_col(alpha = 0.7)
 + ggtitle('Total COVID Cases by State')
 + theme_minimal()
 + labs(x = 'Date', y = 'Total Cases')
 + theme(axis_text_x = element_text(rotation=90),
         plot_title = element_text(size=18),
        figure_size = (12,8))
 + facet_wrap('~state'))
```

## Total population by state

Total population by state

```python
# group and summarise with tp
adm1_sel = (
    tp.from_pandas(nga_admpop_1)
    .select('ADM1_NAME', 'T_TL')
    .summarize(T_TL = tp.sum(col('T_TL')),
               by = 'ADM1_NAME')
).to_pandas()

# group and summarise with tp
adm1_ttl = (
    tp.from_pandas(nga_admpop_1)
    .select('ADM1_NAME', 'T_TL')
    .summarize(T_TL = tp.sum(col('T_TL')),
               by = 'ADM1_NAME')
).to_pandas()

adm1 = adm1_sel.copy()

# Note that value_counts() is already sorted
state_list = adm1.sort_values('T_TL')['ADM1_NAME'].tolist()
state_cat = pd.Categorical(adm1['ADM1_NAME'], categories=state_list)

# assign to a new column in the DataFrame
adm1 = adm1.assign(state_cat = state_cat)
```

```python
# group and summarise with tp
top_cities = (
    tp.from_pandas(nga_admpop_2)
    .select('ADM2_NAME', 'ADM1_NAME', 'T_TL')
    .summarize(T_TL = tp.sum(col('T_TL')),
               by = ['ADM2_NAME', 'ADM1_NAME'])
).to_pandas().rename(columns = {'ADM2_NAME':'city', 'ADM1_NAME':'state', 'T_TL':'population'})

top_cities_tbl = top_cities.sort_values('population', ascending=False).head(10)

# Note that value_counts() is already sorted
city_list = top_cities_tbl.sort_values('population')['city'].tolist()
city_cat = pd.Categorical(top_cities_tbl['city'], categories=city_list)

# assign to a new column in the DataFrame
top_cities_tbl = top_cities_tbl.assign(city_cat = city_cat)
```

```python
(ggplot(adm1, aes('state_cat', 'T_TL'))
 + geom_col(aes(fill = 'T_TL'))
 + ggtitle('Total population by state')
 + theme_minimal()
 + labs(x = 'State', y = 'Population', fill = 'Population')
 + coord_flip()
 + theme(axis_text_x = element_text(rotation=90),
         plot_title = element_text(size=18),
        figure_size = (6,12)))
```

We can see from the data that Lagos, Kano and Federal Capital Territory are the top 3 states for COVID cases. 

#### 1. Lagos

Lagos is the most populous state in Nigeria, and is home to 5 of Nigeria's largest cities by population. It may be unsurprising that it also ranks highest for COVID cases. We can see below though that while it homes 7.3% of the population, it accounts for 43.1% of Nigeria's COVID cases, confirming that population density has a large impact on the spread of COVID, with over 18 cases per 100,000 people in this first few weeks of the pandemic.

#### 2. Kano

This theory is supported by Kano's COVID rate - it is botht he 2nd most populous and highest for COVID cases, but the case rate is much lower at 6.2 cases per 100,000. Kano's largest city Nassawara is only the 7th largest in the country. This implies Kano may be less densely populated than Lagos and Federal Capital Territory.

#### 3. Federal Capital Territory

Federal Capital Territory provides more evidence for this. Despite being the 19th most populous state, it is home to 2 of the country's largest cities, including the 2nd biggest by population, Abuja. These densely populated citiesmay provide the answer to why this smaller state has a case rate of 8.8 per 100,000 and the 3rd highest number of cases nationally.

Combining state cases, population and location data

```python
# group and summarise with tp
adm1_plot = (
    tp.from_pandas(nga_admpop_1)
    .select('ADM1_PCODE', 'T_TL')
    .summarize(T_TL = tp.sum(col('T_TL')),
               by = 'ADM1_PCODE')
)

adm1_plot = adm1_plot.to_pandas()
adm1_plot.rename(columns = {'ADM1_PCODE':'admin1Pcode', 'T_TL':'population'}, inplace = True)

# create adm1 lookup
adm_lkp = (
    tp.from_pandas(nga_admpop_1)
    .select('ADM1_NAME', 'ADM1_PCODE')
)

adm_lkp = adm_lkp.to_pandas()
```

```python
# group and summarise with tp
state_case_sel = (
    tp.from_pandas(state_case_dates)
    .select('date', 'province/state', 'total_cases')
)

state_case_sel = state_case_sel.to_pandas()
state_case_sel['ADM1_NAME'] = state_case_sel['province/state'].str.upper()
state_case_sel = state_case_sel.drop('province/state', axis=1)
state_case_sel['date'] = pd.to_datetime(state_case_sel['date'])

# state_case_sel.head()
```

```python
state_cases = state_case_sel.merge(adm_lkp, on = "ADM1_NAME", how='left').drop('ADM1_NAME', axis=1)

state_cases.rename(columns = {'ADM1_PCODE':'admin1Pcode'}, inplace = True)
```

## Animated plot of cases

```python
from plotnine.animation import PlotnineAnimation
from matplotlib import animation

# for animation in the notebook
from matplotlib import rc
rc('animation', html='html5')

# Parameters used to control the spiral
n = 82
kmin = 0
kmax = 82
num_frames = 164

# geopandas included map, filtered to just Nigeria
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.crs = "EPSG:4326"
nigeria = world[(world['name'] == 'Nigeria')]
# base = nigeria.plot(color='white', edgecolor='black', alpha=0, figsize=(11, 11))

def plot(k):
    # For every plot we change the theta
    state_cases['date'] = pd.to_datetime(state_cases['date'])
    
    _date = min(state_cases['date']) + pd.DateOffset(k.round(decimals=0))
    
    states = state_cases[state_cases['date'] <= pd.to_datetime(_date)]
    
    states_sum = (
    tp.from_pandas(states)
    .summarize(total_cases = tp.sum(col('total_cases')),
               by='admin1Pcode')
        )
    
    states_sum = states_sum.to_pandas()
    
    layers = fiona.listlayers(states_file)
    zones_gdf = gpd.read_file(states_file, layer=1)
    zones_gdf = zones_gdf.merge(adm1_plot, on='admin1Pcode', how='left')
    zones_gdf = zones_gdf.merge(states_sum, on='admin1Pcode', how='left')
    zones_gdf['case_rate'] = zones_gdf['total_cases'] / zones_gdf['population']
    zones_gdf.crs = "EPSG:4326"
    zones_gdf = zones_gdf.set_index('admin1Name_en')

    nga_anim = zones_gdf.copy()

    p = (ggplot(nga_anim)
 + geom_map(nga_anim, aes(fill = 'total_cases'))
 + ggtitle('Total COVID Cases by state')
 + scale_fill_continuous(limits=(0, 500))
 + labs(fill = 'Total_Cases')
 + theme_minimal()
 + coord_fixed()
 + theme(axis_text = element_blank(),
         panel_grid = element_blank(),
         plot_title = element_text(size=24),
        figure_size = (10,12)))
    
    return p

# It is better to use a generator instead of a list
plots = (plot(k) for k in np.linspace(kmin, kmax, num_frames))
ani = PlotnineAnimation(plots, interval=100, repeat_delay=500)
ani.save('animation.gif')
#ani

# writergif = animation.PillowWriter(fps=30)
# ani.save('animation.gif',writer=writergif)
```

```python
# group and summarise with tp
cases_split = (
    tp.from_pandas(cases_df)
    .mutate(date_check = col('date') > pd.to_datetime('2020-04-12'))
    .mutate(wave = col('date_check') + 1)
)

cases_split = cases_split.to_pandas()

cases_split['week'] = cases_split['date'].dt.week
cases_split['week'] = cases_split['week'] - min(cases_split['week']) + 1

wave_lkp = (
    tp.from_pandas(cases_split)
    .summarize(wave = tp.mean('wave'),
               week = tp.mean('week'),
               by = 'date')
).to_pandas()
```

The cases by date chart created earlier shows a clear change in trajectory of case numbers from 12th April 2020. We'll treat cases before this date as 'wave 1' and cases after as 'wave 2' and find out if these spikes were caused by cases increasing exponentially in one state first, and then spreading, or all states simulataneously.

```python
(ggplot(cases_split, aes('date', 'total_cases'))
 + geom_col(aes(fill = 'wave'))
 + geom_smooth()
 + ggtitle('Total Cases by Date')
 + theme_minimal()
 +labs(x = 'Date', y = 'Total Cases', fill = 'Wave')
 + scale_fill_continuous(guide=False)
 + theme(axis_text_x = element_text(rotation=90),
         plot_title = element_text(size=18),
        figure_size = (12, 5)))
```

## Wave 1 by state

```python
state_waves = state_cases.merge(wave_lkp, on = 'date', how = 'left')

wave_1 = state_waves[state_waves['wave'] == 1]

wave_1 = (
    tp.from_pandas(wave_1)
    .summarize(total_cases = tp.sum(col('total_cases')),
              by = 'admin1Pcode')
).to_pandas()

zones_gdf_w1 = gpd.read_file(states_file, layer=1)
zones_gdf_w1 = zones_gdf_w1.merge(adm1_plot, on='admin1Pcode', how='left')
zones_gdf_w1 = zones_gdf_w1.merge(wave_1, on='admin1Pcode', how='left')
zones_gdf_w1['case_rate'] = zones_gdf_w1['total_cases'] / zones_gdf_w1['population'] * 100000
zones_gdf_w1.crs = "EPSG:4326"
zones_gdf_w1 = zones_gdf_w1.set_index('admin1Name_en')

```

```python
(ggplot(zones_gdf_w1)
 + geom_map(zones_gdf_w1, aes(fill = 'case_rate'))
 + ggtitle('Rate of COVID cases in Nigeria by state during initial wave')
 + labs(fill = 'Cases per\n100,000 people')
 + theme_minimal()
 + coord_fixed()
 + theme(axis_text = element_blank(),
         panel_grid = element_blank(),
         legend_title = element_text(margin = {'b':20}),
         plot_title = element_text(size=24),
         figure_size = (10, 12)))
```

## Wave 2

```python
wave_2 = state_waves[state_waves['wave'] == 2]

wave_2 = (
    tp.from_pandas(wave_2)
    .summarize(total_cases = tp.sum(col('total_cases')),
              by = 'admin1Pcode')
).to_pandas()

zones_gdf_w2 = gpd.read_file(states_file, layer=1)
zones_gdf_w2 = zones_gdf_w2.merge(adm1_plot, on='admin1Pcode', how='left')
zones_gdf_w2 = zones_gdf_w2.merge(wave_2, on='admin1Pcode', how='left')
zones_gdf_w2['case_rate'] = zones_gdf_w2['total_cases'] / zones_gdf_w2['population'] * 100000
zones_gdf_w2.crs = "EPSG:4326"
zones_gdf_w2 = zones_gdf_w2.set_index('admin1Name_en')
```
